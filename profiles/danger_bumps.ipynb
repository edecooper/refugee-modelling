```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import quad
import os
import nbimporter
```

```python
fig = plt.figure()
ax = fig.add_subplot()
colors = sns.color_palette("crest", 5)
ax.set_prop_cycle('color', colors)
bins = np.linspace(0.1,1.1,5)
ts = np.arange(0,3*10**4)
dt = 1e-3
for i in bins: ax.plot(ts*dt,  i/(1 + (ts*dt - 3)**2), label = "$P_k = " + str(i) + "$", linewidth = 3)
ax.legend()
ax.set_xlabel("Time (days?)")
ax.set_ylabel("Danger")
fig.savefig("plots/single_bump.png", dpi = 300)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_prop_cycle('color', colors)
for i in bins: ax.plot(ts*dt,  0.5*i/(1 + (ts*dt - 3)**2) + 0.5*i/(1 + (ts*dt - 7)**2), label = "$P_k = " + str(i) + "$", linewidth = 3)
ax.legend()
ax.set_xlabel("Time ($t$)")
ax.set_ylabel("Danger")
fig.savefig("plots/double_bump.png", dpi = 300)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
colors = sns.color_palette("crest", 5)
ax.set_prop_cycle('color', colors)
bins = np.linspace(0.1,1.1,5)
ts = np.arange(0,3*10**4)
dt = 1e-3
for i in bins: ax.plot(ts*dt,  ts*dt + np.sin(ts*dt*3)/3 +  np.sin(ts*dt*7)/3 + np.sin(ts*dt*2)/3, linewidth = 3, color = colors[-1])
ax.set_xlabel("Time (years?)")
ax.set_ylabel("Danger")
fig.savefig("plots/irish.png", dpi = 300)
```
